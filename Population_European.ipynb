```python
# Importing the libraries

import pandas as pd
import numpy as np
import math as mt
import seaborn as sns #visualisation
import matplotlib.pyplot as plt # visualisation
plt.style.use("seaborn-whitegrid")
import warnings            
warnings.filterwarnings("ignore") # We can suppress the warnings
import statistics as stats #calculation
```

```python
# Importing the dataset Price

dfpop5 =pd.read_csv("1.Population_Famale.csv")
dfpop6 =pd.read_csv("2.Population_Male.csv")
dfpop7 =pd.read_csv("3.Population_Rural.csv")
dfpop8 =pd.read_csv("4.Population_Urban.csv")
```

```python
# change names on datasets

dfpop5 = dfpop5.rename(columns = {"Value":"Famale_K "})
dfpop6 = dfpop6.rename(columns = {"Value":"Male_K"})
dfpop7 = dfpop7.rename(columns = {"Value":"Rural_K"})
dfpop8 = dfpop8.rename(columns = {"Value":"Urban_K"})
```

```python
#Created a join tables
dfpop_m3 = pd.merge(dfpop5,dfpop6, how='inner', on=["Area","Year"],suffixes=("_dfpop5", "_dfpop6"))
dfpop_m4 = pd.merge(dfpop7,dfpop8, how='inner', on=["Area","Year"],suffixes=("_dfpop7", "_dfpop8"))
df_pop = pd.merge(dfpop_m3,dfpop_m4, how='inner', on=["Area","Year"])
```

```python
#full print
df_pop
```

```python
# Clean the dataset, deleting useless columns for ou analyse

df_pop.drop(["Domain_Code_dfpop5","Domain_Code_dfpop6","Domain_Code_dfpop7","Domain_Code_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Domain_dfpop5","Domain_dfpop6","Domain_dfpop7","Domain_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Area_Code_(M49)_dfpop5","Area_Code_(M49)_dfpop6","Area_Code_(M49)_dfpop7","Area_Code_(M49)_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Element_Code_dfpop5","Element_Code_dfpop6","Element_Code_dfpop7","Element_Code_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Element_dfpop5","Element_dfpop6","Element_dfpop7","Element_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Item_Code_dfpop5","Item_Code_dfpop6","Item_Code_dfpop7","Item_Code_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Item_dfpop5","Item_dfpop6","Item_dfpop7","Item_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Year_Code_dfpop5","Year_Code_dfpop6","Year_Code_dfpop7","Year_Code_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Unit_dfpop5","Unit_dfpop6","Unit_dfpop7","Unit_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Flag_dfpop5","Flag_dfpop6","Flag_dfpop7","Flag_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Flag_Description_dfpop5","Flag_Description_dfpop6","Flag_Description_dfpop7","Flag_Description_dfpop8"], axis=1, inplace=True)
df_pop.drop(["Note_dfpop5","Note_dfpop6","Note_dfpop7","Note_dfpop8"], axis=1, inplace=True)

# Prints out the first few rows
print(df_pop.head())
```

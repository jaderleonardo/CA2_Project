```python
# Importing the libraries

import pandas as pd
import numpy as np
import math as mt
import seaborn as sns #visualisation
import matplotlib.pyplot as plt # visualisation
plt.style.use("seaborn-whitegrid")
import warnings            
warnings.filterwarnings("ignore") # We can suppress the warnings
import statistics as stats #calculation
```

```python
# Importing the dataset Production

df5 =pd.read_csv("5.Barley_Price.csv")
df6 =pd.read_csv("6.Meat_Cattle_Price.csv")
df7 =pd.read_csv("7.Meat_Pig_Price.csv")
df8 =pd.read_csv("8.Raw_Milk_Price.csv")
```

```python
# change names on datasets

df5 = df5.rename(columns = {"Value":"Barley"})
df6 = df6.rename(columns = {"Value":"Meat Cattle"})
df7 = df7.rename(columns = {"Value":"Meat Pig"})
df8 = df8.rename(columns = {"Value":"Raw Milk"})
```

```python
#Created a join tables
df_m3 = pd.merge(df5,df6, how='inner', on=["Area","Year"],suffixes=("_df5", "_df6"))
df_m4 = pd.merge(df7,df8, how='inner', on=["Area","Year"],suffixes=("_df7", "_df8"))
df_price = pd.merge(df_m3,df_m4, how='inner', on=["Area","Year"])
```

```python
#full print
df_price
```

```python
# Clean the dataset, deleting useless columns for ou analyse

df_price.drop(["Domain_Code_df5","Domain_Code_df6","Domain_Code_df7","Domain_Code_df8"], axis=1, inplace=True)
df_price.drop(["Domain_df5","Domain_df6","Domain_df7","Domain_df8"], axis=1, inplace=True)
df_price.drop(["Area_Code_(M49)_df5","Area_Code_(M49)_df6","Area_Code_(M49)_df7","Area_Code_(M49)_df8"], axis=1, inplace=True)
df_price.drop(["Element_Code_df5","Element_Code_df6","Element_Code_df7","Element_Code_df8"], axis=1, inplace=True)
df_price.drop(["Element_df5","Element_df6","Element_df7","Element_df8"], axis=1, inplace=True)
df_price.drop(["Item_Code_(CPC)_df5","Item_Code_(CPC)_df6","Item_Code_(CPC)_df7","Item_Code_(CPC)_df8"], axis=1, inplace=True)
df_price.drop(["Item_df5","Item_df6","Item_df7","Item_df8"], axis=1, inplace=True)
df_price.drop(["Year_Code_df5","Year_Code_df6","Year_Code_df7","Year_Code_df8"], axis=1, inplace=True)
df_price.drop(["Months_Code_df5","Months_Code_df6","Months_Code_df7","Months_Code_df8"], axis=1, inplace=True)
df_price.drop(["Months_df5","Months_df6","Months_df7","Months_df8"], axis=1, inplace=True)
df_price.drop(["Unit_df5","Unit_df6","Unit_df7","Unit_df8"], axis=1, inplace=True)
df_price.drop(["Flag_df5","Flag_df6","Flag_df7","Flag_df8"], axis=1, inplace=True)
df_price.drop(["Flag_Description_df5","Flag_Description_df6","Flag_Description_df7","Flag_Description_df8"], axis=1, inplace=True)

# Prints out the first few rows
print(df_price.head())
```
